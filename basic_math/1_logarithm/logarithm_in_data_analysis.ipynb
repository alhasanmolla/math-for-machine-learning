```python
import numpy as np
import pandas as pd
```

```python
# 1. Create the dataset exactly as shown in the image
data = {
    "person name": ["Rob", "Tom", "Xi", "Mohan", "Pooja", "Sofiya"],
    "credit score": [750, 310, 475, 600, 820, 780],
    "income": [80000, 32000, 77000, 65000, 550000, 75000],
    "age": [32, 45, 33, 51, 35, 31],
    "loan approved?": ["Y", "N", "Y", "N", "Y", "Y"],
}

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.plot(x='person name', y='income', kind='bar')
```

```python
# 2. Calculate the base-10 logarithm of the income column
df["log income"] = np.log10(df["income"])
df.head()
```

```python
df.plot(x='person name', y='log income', kind='bar')
```

```python
df['_income'] = 10 ** df['log income']
df.head()
```

```python
df.plot(x='person name', y='_income', kind='bar')
```
